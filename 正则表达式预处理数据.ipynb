```python
import os
import re
import pandas as pd

# 定义正则表达式
#title_pattern = r'^2017-08.*\.txt$'
title_pattern = r'^20.*\.txt$'
start_pattern = r'^(.*事故发生经过.*)$'
emergency_pattern = r'^(.*应急救援情况.*|.*应急处置情况.*)$'
direct_causes_pattern = r'^(.*直接原因.*)$'
indirect_causes_pattern = r'^(.*间接原因.*)$'
responsibility_pattern = r'^(.*建议追究责任人员.*)$'
improvement_pattern = r'^(.*事故防范和整改措施.*)$'

# 创建空的数据框
df = pd.DataFrame(columns=['文件名', '摘要', '事故发生经过', '应急处置情况', '直接原因', '间接原因', '建议追究责任人员', '事故防范和整改措施'])

# 遍历文件夹中的所有文件
for file_name in os.listdir('D:\\KnowlegdgeGraph\\TextDataset\\三阶数据'):
    file_path = os.path.join('D:\\KnowlegdgeGraph\\TextDataset\\三阶数据', file_name)
    if os.path.isfile(file_path) and re.match(title_pattern, file_name):
        # 如果是以2017-08开头的txt文件，提取信息并添加到数据框中
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            start_found = False 
            emergency_found = False
            direct_causes_found = False
            indirect_causes_found = False
            responsibility_found = False
            improvement_found = False
            # 修改提取“摘要”的代码
            summary_text = ''
            for line in lines:
                if line.strip() == '':
                    # 遇到空白行，停止提取该信息
                    break
                summary_text += line
            #精确匹配
            #summary_text = lines[0].strip()
            start_text = ''
            emergency_text = ''
            direct_causes_text = ''
            indirect_causes_text = ''
            responsibility_text = ''
            improvement_text = ''
            for line in lines:
                if not start_found and re.search(start_pattern, line):
                    start_found = True
                elif start_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    start_found = False
                    emergency_found = False
                    direct_causes_found = False
                    indirect_causes_found = False
                    responsibility_found = False
                    improvement_found = False
                elif start_found and not line.isspace():
                    start_text += line
            for line in lines:
                if not emergency_found and re.search(emergency_pattern, line):
                    emergency_found = True
                elif emergency_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    emergency_found = False
                    direct_causes_found = False
                    indirect_causes_found = False
                    responsibility_found = False
                    improvement_found = False
                elif emergency_found and not line.isspace():
                    emergency_text += line
            for line in lines:    
                if not direct_causes_found and re.search(direct_causes_pattern, line):
                    direct_causes_found = True
                elif direct_causes_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    direct_causes_found = False
                    indirect_causes_found = False
                    responsibility_found = False
                    improvement_found = False
                elif direct_causes_found and not line.isspace():
                    direct_causes_text += line
            for line in lines: 
                if not indirect_causes_found and re.search(indirect_causes_pattern, line):
                    indirect_causes_found = True
                elif indirect_causes_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    indirect_causes_found = False
                    responsibility_found = False
                    improvement_found = False
                elif indirect_causes_found and not line.isspace():
                    indirect_causes_text += line
            for line in lines:
                if not responsibility_found and re.search(responsibility_pattern, line):
                    responsibility_found = True
                elif responsibility_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    responsibility_found = False
                    improvement_found = False
                elif responsibility_found and not line.isspace():
                    responsibility_text += line
            for line in lines:
                if not improvement_found and re.search(improvement_pattern, line):
                    improvement_found = True
                elif improvement_found and line.isspace():
                    # 遇到空白行，停止提取该信息
                    improvement_found = False
                elif improvement_found and not line.isspace():
                    improvement_text += line
            
            # 将提取到的信息添加到数据框中
            df = df.append({
                '文件名': file_name,
                '摘要': summary_text,
                '事故发生经过': start_text,
                '应急处置情况': emergency_text,
                '直接原因': direct_causes_text,
                '间接原因': indirect_causes_text,
                '建议追究责任人员': responsibility_text,
                '事故防范和整改措施': improvement_text
            }, ignore_index=True)

# 将数据框保存到 Excel 表格中
df.to_excel('D:\\KnowlegdgeGraph\\TextDataset\\测试结果.xlsx', index=False)
```
